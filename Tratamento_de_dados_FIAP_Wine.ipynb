```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
pd.options.display.float_format = "{:.2f}".format
```

```python
base_exportacao = pd.read_csv("https://raw.githubusercontent.com/guipereira42/3DTAT_TC_FIAP_01/main/dados/ExpVinho.csv", encoding = "utf-8", sep = ";")
base_exportacao.head()
```

```python
base_exportacao.drop(base_exportacao.columns[0], axis=1, inplace = True)
base_exportacao.head()
```

```python
def separar_colunas(ini, base_dados, coluna_id):
  ini = ini
  headers = [coluna_id]
  for cont in range(int(len(base_dados.columns[1:].tolist())/2)):
    headers.append(base_dados.columns[ini])
    ini = ini + 2
  return headers

base_exportacao_quantidade = base_exportacao[separar_colunas(1, base_exportacao, base_exportacao.columns[0])]
base_exportacao_valor = base_exportacao[separar_colunas(2, base_exportacao, base_exportacao.columns[0])]
```

```python
base_exportacao_quantidade = base_exportacao_quantidade.melt(id_vars= base_exportacao_quantidade.columns[0], value_vars = base_exportacao_quantidade.columns[1:])
base_exportacao_quantidade.columns = ['pais', 'ano', 'quantidade_l']
base_exportacao_quantidade
```

```python
base_exportacao_valor = base_exportacao_valor.melt(id_vars= base_exportacao_valor.columns[0], value_vars = base_exportacao_valor.columns[1:])
base_exportacao_valor.columns = ['pais', 'ano', 'valor_dol']
base_exportacao_valor['ano'] = base_exportacao_valor['ano'].str[:4]
base_exportacao_valor
```

```python
base_exportacao_em_colunas = pd.merge(base_exportacao_quantidade, base_exportacao_valor, on=['pais', 'ano'])
base_exportacao_em_colunas
base_exportacao_em_colunas.to_excel("Exportacao_Qtd_Vlr.xlsx")
```

```python
base_producao = pd.read_csv("https://raw.githubusercontent.com/guipereira42/3DTAT_TC_FIAP_01/main/dados/Producao.csv", encoding = "utf-8", sep = ";")
base_producao.drop("id", axis=1, inplace = True)
base_producao.head(15)
```

```python
base_producao['produto'].to_list()
```

```python
base_producao['tipo'] = base_producao['produto']
```

```python
base_producao['tipo'][:4] = 'VINHO DE MESA'
base_producao['tipo'][4:8] = 'VINHO FINO DE MESA'
base_producao['tipo'][8:14] = 'SUCO'
base_producao['tipo'][14:] = 'DERIVADOS'
base_producao.drop([0, 4, 8, 14], inplace=True)

base_producao
```

```python
base_producao_em_colunas = base_producao.melt(id_vars= ['produto','tipo'] , value_vars = base_producao.columns[1:-1])
base_producao_em_colunas.columns = ['produto', 'tipo', 'ano', 'quantidade_l']
base_producao_em_colunas
base_producao_em_colunas.to_excel("Tipo_Producao_Uva.xlsx")
```
